#1. Análisis de facturación



```python
ingreso_total_tienda1 = tienda["Precio"].sum()
ingreso_total_tienda2 = tienda2["Precio"].sum()
ingreso_total_tienda3 = tienda3["Precio"].sum()
ingreso_total_tienda4 = tienda4["Precio"].sum()

print(ingreso_total_tienda1)
print(ingreso_total_tienda2)
print(ingreso_total_tienda3)
print(ingreso_total_tienda4)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.barh(y=["Tienda1", "Tienda2", "Tienda3", "Tienda4"], width=[ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4], color=["red", "pink", "yellow", "green"])
plt.xlabel('Ingreso')
plt.ylabel('Tienda')
plt.title('Ingreso por tienda')

# Establece los límites del eje X (ingresos)
# Reemplaza 'valor_minimo_deseado' y 'valor_maximo_deseado' con los límites que quieras
plt.xlim(left=0, right=1200000000)
plt.savefig("grafico1de1.png")
plt.show()
```

```python
plt.pie(x=[ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4], labels=["Tienda1", "Tienda2", "Tienda3", "Tienda4"], autopct='%1.2f%%', colors=["red", "pink", "yellow", "green"])
plt.title("Porcentaje de ingresos por tienda\n la tienda 4 es la que menos ingresos tiene")
plt.show()
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
ventas_por_categoria_tienda1 = tienda.groupby("Categoría del Producto")["Categoría del Producto"].count()
ventas_por_categoria_tienda2 = tienda2.groupby("Categoría del Producto")["Categoría del Producto"].count()
ventas_por_categoria_tienda3 = tienda3.groupby("Categoría del Producto")["Categoría del Producto"].count()
ventas_por_categoria_tienda4 = tienda4.groupby("Categoría del Producto")["Categoría del Producto"].count()
print(ventas_por_categoria_tienda1)
print(ventas_por_categoria_tienda2)
print(ventas_por_categoria_tienda3)
print(ventas_por_categoria_tienda4)
```

```python
ventas_por_categoria_total = pd.DataFrame({
    'Tienda1': ventas_por_categoria_tienda1,
    'Tienda2': ventas_por_categoria_tienda2,
    'Tienda3': ventas_por_categoria_tienda3,
    'Tienda4': ventas_por_categoria_tienda4
})

print(ventas_por_categoria_total.head(20))
```

```python
ventas_por_categoria_total.plot(kind='bar', figsize=(12, 7))
plt.title('Ventas por Categoría de Producto por Tienda de Alura Store')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda')
plt.tight_layout()
plt.savefig("grafico1de2.png")
plt.show()
```

```python
ventas_por_categoria_total.plot(kind='line', figsize=(12, 7), marker='o')
plt.title('Ventas por Categoría de Producto por Tienda Alura Store (Gráfico de Líneas)')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tiendas')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
valor_promedio_envio_tienda1 = tienda["Costo de envío"].mean().round(2)
valor_promedio_envio_tienda2 = tienda2["Costo de envío"].mean().round(2)
valor_promedio_envio_tienda3 = tienda3["Costo de envío"].mean().round(2)
valor_promedio_envio_tienda4 = tienda4["Costo de envío"].mean().round(2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(
    x=["Tienda1", "Tienda2", "Tienda3", "Tienda4"],
    height=[valor_promedio_envio_tienda1,valor_promedio_envio_tienda2,valor_promedio_envio_tienda3,valor_promedio_envio_tienda4
    ]
)
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.title('Costo de Envío Promedio por Tienda')
plt.savefig("grafico1de5.png")
plt.show()
```

### Gráfico de Líneas con Marcadores para el Costo de Envío Promedio por Tienda

Este gráfico representa el costo de envío promedio de cada tienda como puntos conectados por líneas, ofreciendo una alternativa a los gráficos de barras y circulares para comparar estos valores.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(
    ["Tienda1", "Tienda2", "Tienda3", "Tienda4"],
    [valor_promedio_envio_tienda1, valor_promedio_envio_tienda2, valor_promedio_envio_tienda3, valor_promedio_envio_tienda4],
    marker='o', linestyle='-', color='gray'
)
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.title('Costo de Envío Promedio por Tienda')
plt.savefig("grafico2de5.png")
plt.show()
```
